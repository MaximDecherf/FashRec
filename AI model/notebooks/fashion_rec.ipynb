### Encoding of mult cat var

```python
df_encoded = pd.get_dummies(fashion_df, columns=["gender", "masterCategory","subCategory","articleType","baseColour","season", "usage" ])
df_encoded.drop(['productDisplayName'], axis = 1, inplace = True) 
df_encoded.dropna(inplace=True)
df_encoded.columns
```

### Data from cold start User 

```python
likes = df_encoded[:10]
dislikes = df_encoded[20:30]
```

```python
likes["likes"] = True
dislikes["likes"] = False
```

### Decision Tree

```python
preferences_df = pd.concat([likes, dislikes], axis=0)
```

```python
preferences_df
```

```python
# cols to train on (without id and like(target class))
preferences_df.columns[1:-1] 
```

```python
features =  preferences_df.columns[1:-1]

X_train = preferences_df[features]
y_train = preferences_df["likes"]
```

```python
from sklearn.tree import DecisionTreeClassifier


clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
# y_pred = clf.predict(X_train.iloc[50].array.reshape(-1, 1).to_numpy().flatten())
```
